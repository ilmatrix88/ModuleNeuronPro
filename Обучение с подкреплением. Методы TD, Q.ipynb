```python
import gym
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# Создание среды Taxi-v3
env = gym.make("Taxi-v3")

# Параметры для поиска по сетке
gamma_values = [0.9, 0.95, 0.99]
alpha_values = [0.1, 0.2, 0.3]
epsilon_values = [0.1, 0.2, 0.3]

# Количество эпизодов для обучения
num_episodes = 1000

# Функция для выполнения эпизода
def run_episode(env, Q1, Q2, alpha, gamma, epsilon):
    state = env.reset()
    total_reward = 0
    done = False
    while not done:
        if np.random.rand() < epsilon:
            action = env.action_space.sample()  # Случайное действие
        else:
            action = np.argmax(Q1[state] + Q2[state])  # Действие по стратегии

        next_state, reward, done, _ = env.step(action)
        total_reward += reward

        if np.random.rand() < 0.5:
            best_next_action = np.argmax(Q1[next_state])
            td_target = reward + gamma * Q2[next_state][best_next_action]
            td_error = td_target - Q1[state][action]
            Q1[state][action] += alpha * td_error
        else:
            best_next_action = np.argmax(Q2[next_state])
            td_target = reward + gamma * Q1[next_state][best_next_action]
            td_error = td_target - Q2[state][action]
            Q2[state][action] += alpha * td_error

        state = next_state

    return total_reward

# Функция для обучения с двойным Q-обучением
def train_double_q_learning(env, gamma, alpha, epsilon, num_episodes):
    Q1 = np.zeros((env.observation_space.n, env.action_space.n))
    Q2 = np.zeros((env.observation_space.n, env.action_space.n))
    total_rewards = []

    for episode in tqdm(range(num_episodes)):
        total_reward = run_episode(env, Q1, Q2, alpha, gamma, epsilon)
        total_rewards.append(total_reward)

    return total_rewards

# Поиск по сетке параметров
results = {}

for gamma in gamma_values:
    for alpha in alpha_values:
        for epsilon in epsilon_values:
            print(f"Training with gamma={gamma}, alpha={alpha}, epsilon={epsilon}")
            total_rewards = train_double_q_learning(env, gamma, alpha, epsilon, num_episodes)
            avg_reward = np.mean(total_rewards)
            results[(gamma, alpha, epsilon)] = avg_reward

# Вывод результатов
for params, avg_reward in results.items():
    print(f"Params: {params}, Avg Reward: {avg_reward}")

# Построение графиков
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

for i, gamma in enumerate(gamma_values):
    for j, alpha in enumerate(alpha_values):
        ax = axs[i, j]
        epsilon_rewards = [results[(gamma, alpha, epsilon)] for epsilon in epsilon_values]
        ax.plot(epsilon_values, epsilon_rewards, marker='o')
        ax.set_title(f"Gamma={gamma}, Alpha={alpha}")
        ax.set_xlabel("Epsilon")
        ax.set_ylabel("Avg Reward")

plt.tight_layout()
plt.show()
```
